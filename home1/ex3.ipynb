```julia
using Plots
using LaTeXStrings
```

```julia
n=50;
h=1/(n+1)
x=LinRange(0,1,n);
```

```julia
b=zeros(n);
bb=zeros(n);
A=zeros(n,n);
f(i)=-pi*(2*i*cos(pi*i*h)+(1-i)*cos(pi*(i-1)*h)-(1+i)*cos(pi*(i+1)*h))+(sin(pi*(i-1)*h)+sin(pi*(i+1)*h))-pi*((i-1)*h*cos(pi*(i-1))+(i+1)*h*cos(pi*(i+1)));
#function for the excersice a)
ff(i)=(-1/h)*(exp((i+1)*h)+exp((i-1)*h)-2*exp(i*h))
#function for the excersice b)
w(x)=1+x*(exp(1)-1);
splat(f) = x -> f(x...);
```

```julia

for i in 1:n
    A[i,i]=2/h;
    if i < n  
        A[i+1,i]=-1/h;
        A[i,i+1]=-1/h;
        if i > 1
            b[i]=f(i) #for 3a)
            bb[i]=ff(i) #for 3B)  
        end
    end

end 
```

```julia
c=(A^-1)*b;
c[1]=0;
c[end]=0;
```

```julia
plot(x,c, label="-Δu=π^2sin(πx)")
```

```julia

cc=(A^-1)*bb;
cc[1]=0;
cc[end]=0;
```

```julia
plot(x,cc, label="-Δv=exp(x)")
plot!(x,cc+splat(w).(x), label="u=v+w")
```
